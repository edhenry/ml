## Probability Theory

* Helps quantify uncertainty
    * Noise on measurements
    * Finite size of a given dataset
    
#### Two box example :
One red box, and one blue box. Within these two boxes we have a mixture of fruit.

* Red Box
    * 2 Apples
    * 6 Oranges

* Blue Box
    * 3 Apple
    * 1 Orange
    


(image courtest of 1)

Random Variables : 

* $B$ - box color
    * $B_{r}$ - red box
    * $B_{b}$ - blue box
* $F$ - fruit type
    * $F_{a}$ - apple
    * $F_{o}$ - orange
    
Definition of the probability of an event to be :

$$ \text{probability of event} = \frac{\text{times a specific event occurs}}{\text{total number of events}}$$

##### Assumptions

* Probability of selecting a red box :
$$p(B = r) = \frac{4}{10}$$

* Probability of selecting a blue box : 

$$P(B = b) = \frac{6}{10}$$

* Note by definitions, probabilities must lie within the interval $[0,1]$
* Also if events are mutually exclusive and include all possible outcomes then we see that the probabilities of the events must sum to $1$.

#### Sum and Product rules of Probability

* More general example than that of the colored boxes and fruit

Random Variables : 

* $X$
    * $X$ can take any value $x_{i}$ where $i = 1,...,M$
* $Y$
    * $Y$ can take any value $y_{j}$ where $j = 1,...,L$
* $N$ (total number of trials where we sample both variables)
    * $n_{ij} = (X = x_{i}, Y = y_{j})$
    * $c_{i}$ = $(X = x_{i}$ irrespective of $Y$)
    * $r_{j}$ = $(Y = y_{j}$ irrespective of $X$)

The probability that $X = x_{i}$ and $Y = y_{j}$ is written as $p(X = x_{i}, Y = y_{j})$, otherwise known as the **joint probability** :

$$ p(X = x_{i}, Y = y_{j}) = \frac{n_{ij}}{N} \ \text{as} \ N \to \infty $$

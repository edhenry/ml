```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

from typing import List, Dict

from matplotlib import pyplot

%matplotlib inline

def remove_str_cols(dataframe):
    cols_to_remove = []

    for col in dataframe.columns:
        try:
            _ = dataframe[col].astype(float)
        except ValueError:
            cols_to_remove.append(col)
            pass
        
    dataframe = dataframe[[col for col in dataframe.columns if col not in cols_to_remove]]     
    return dataframe

def normalize_df(dataframe):
    vals = dataframe.values
    min_max_scaler = preprocessing.MinMaxScaler()
    vals_scaled = min_max_scaler.fit_transform(vals)
    df = pd.DataFrame(vals_scaled)
    df.columns = dataframe.columns
    return df
```

```python
kiran_lab = pd.read_csv('Kiran_lab_SSD_FIO-10-05.csv')

# remove categorical features
kiran_lab_no_str = remove_str_cols(kiran_lab)

# normalize values
vals = kiran_lab_no_str.values
min_max_scaler = preprocessing.MinMaxScaler()
vals_scaled = min_max_scaler.fit_transform(vals)
kiran_lab_norm = pd.DataFrame(vals_scaled)

# re-add column names
kiran_lab_norm.columns = kiran_lab_no_str.columns

# plot some specific columns
kl_plots = pd.DataFrame(kiran_lab_norm, columns=['IOPS', 'Total Throughput (MB\s)', 'disk_write_iops', 
                                                 'write_bw_mean', 'disk_read_iops', 'write_runtime_ms', 
                                                 'read_bandwidth'])
kl_plots.plot(kind='bar', figsize=(15,15), title="10 Node Cluster - Jan 23")
```

```python
single_host_fourk = pd.read_csv("1-host-test on 10 hosts Jan 25.csv", sep=',')

single_host_fourk.dropna(how='all', inplace=True)
single_host_fourk.reset_index(drop=True, inplace=True)
```

```python
def row_slice(dataframe, start_loc, stop_loc):
    return dataframe.loc[start_loc:stop_loc]

counter = 0
start = 0
stop = 4

server_number = 20

dfs = []

while counter <= len(single_host_fourk):
    df = pd.DataFrame(row_slice(single_host_fourk, start, stop))
    df = remove_str_cols(df)
    dfs.append(df)
    start += 5
    stop += 5
    
    counter += 5
```

```python
server_20 = dfs[0]
server_21 = dfs[1]
server_22 = dfs[2]
server_23 = dfs[3]
server_24 = dfs[4]
server_25 = dfs[5]
server_26 = dfs[6]
server_27 = dfs[7]
server_28 = dfs[8]
server_29 = dfs[9]

servers = [server_20, server_21, server_22,
           server_23, server_24, server_25,
           server_26, server_27, server_28,
           server_29]
```

```python
def min_max_mean(dataframes, metric: List) -> pd.DataFrame:
    """
    Grab min, max, and mean stats of each column of dataframe
    
    :param dataframe: Dataframe to grab statistics for
    :param fields: list of fields to grab statistics for
    :return df: dataframe of provided field statistics
    """
    stats = ['min', 'max', 'mean']
    
    server_number = 20
    dfs = {}
    metric_dict = {}
    
    for m in metric:
        all_servers = {}
        for stat in stats:
            agg_stats = []
            for dataframe in dataframes:
                agg_stats.append(dataframe[m].describe()[1:8][stat])
                all_servers["server_%d" % server_number] = agg_stats
            server_number += 1
        dfs[m] = all_servers
    return dfs

def plot_dataframe(dataframe, title: str) -> df.plot:
    
    df = df.rename(index={0: 'min', 1: 'max', 2:'mean'})
    return df.T.plot.bar(figsize=(20,20), title=title)
```
